```python
def bar_chart(features):
    Survived =titanic_train[titanic_train['Survived']==1][features].value_counts()
    dead = titanic_train[titanic_train['Survived']==0][features].value_counts()
    df =pd.DataFrame([Survived,dead])
    df.index = ['Survived','dead']
    df.plot(kind = 'bar',stacked = True,figsize=(10,5))
    
```

```python
features_drop =['Ticket', 'SibSp', 'Parch','Embarked']
titanic_train = titanic_train.drop(features_drop, axis =1)
titanic_test  = titanic_test.drop(features_drop,axis =1)
titanic_train = titanic_train.drop(['PassengerId'],axis =1)
```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits = 10, shuffle = True, random_state =0)
```

```python
classification = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(classification, titanic_train_data, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
round(np.mean(score)*100,2)
```
